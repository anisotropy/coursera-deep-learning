## Sigmoid Function

```python
import math
import matplotlib.pyplot as mpplot
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


xs = np.arange(-10, 10, 0.1)
ys = sigmoid(xs)

mpplot.figure(figsize=(10, 6))
mpplot.grid(True)
mpplot.plot(xs, ys)
```

```python
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)


def sigmoid_derivative_2(x, dx=0.1):
    return (sigmoid(x + dx) - sigmoid(x)) / dx


xs = np.arange(-10, 10, 0.1)
ys = sigmoid_derivative(xs)
ys2 = sigmoid_derivative_2(xs)

mpplot.figure(figsize=(10, 6))
mpplot.grid(True)
mpplot.plot(xs, ys, 'b', label='sigmoid_derivative')
mpplot.plot(xs, ys2, 'r', label='sigmoid_derivative_2')
mpplot.legend()
```

## Image to Vector

```python
def image2vector(image):
    image_shape = image.shape
    reshaped_image = image.reshape(image_shape[0] * image_shape[1] * image_shape[2], 1)
    return reshaped_image

def image2vector_2(image):
    return image.reshape(-1, 1)

image = np.array([
    [[ 0.67826139,  0.29380381],
    [ 0.90714982,  0.52835647],
    [ 0.4215251 ,  0.45017551]],

    [[ 0.92814219,  0.96677647],
    [ 0.85304703,  0.52351845],
    [ 0.19981397,  0.27417313]],

    [[ 0.60659855,  0.00533165],
    [ 0.10820313,  0.49978937],
    [ 0.34144279,  0.94630077]]])


print('image2vector\n', image2vector(image))
print()
print('The result of image2vector() is same as that of image2vector_2()\n', image2vector_2(image))
```

## Normalize

```python
def normalize_row(x):
    x_norm = np.linalg.norm(x, axis=1, keepdims=True)
    normalized = x / x_norm
    return normalized

x = np.array(
    [[0, 3, 4],
    [1, 6, 4]]
)

print(normalize_row(x))
```

## Softmax Function

```python
def softmax_row(x):
    x_exp = np.exp(x)
    x_sum = np.sum(x_exp, axis=1, keepdims=True)
    return x_exp / x_sum


x = np.array([
    [9, 2, 5, 0, 0],
    [7, 5, 0, 0 ,0]])

print(softmax_row(x))
```

## L1 and L2 loss functions

```python
def L1(y_hat, y):
    return np.sum(np.abs(y_hat - y))

yhat = np.array([.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
print("L1 = " + str(L1(yhat,y)))
```

```python
def L2(y_hat, y):
    return np.dot(y_hat - y, y_hat - y)

yhat = np.array([.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
print("L2 = " + str(L2(yhat,y)))
```
